```python
# convert each line into a tuple of date, avg price, total volume, region
def filterSales(sale):
    split_sale = sale.split(',')
    return (split_sale[1], float(split_sale[2]), float(split_sale[3]), split_sale[13])
```

```python
sales_filtered = [filterSales(sale) for sale in sales[1:]]
```

```python
sales_filtered[1]
```

```python
# Now if we want to get just id and avgprice as 'id, avgprice' for each record, we write a new function
def join_demo(sale):
    return ','.join([sale.split(',')[0], sale.split(',')[2]])
```

```python
id_price = [join_demo(sale) for sale in sales[1:]]
```

```python
id_price[:10]
```

```python
# we can also use append function and for loop
id_and_price = []
for sale in sales[1:]:
    id_and_price.append(','.join([sale.split(',')[0], sale.split(',')[2]]))
```

```python
id_and_price[:10]
```

```python
# we can also convert each row into a dict using this function
def get_sales_dict(sale):
    sale_element = sale.split(',')
    return ({
        'date' : sale_element[1], 
        'avg price' : sale_element[2], 
        'total volume' : sale_element[3], 
        'region': sale_element[13]
    })
```

```python
sale_dict = [get_sales_dict(sale) for sale in sales[1:]]
```

```python
sale_dict[:10]
```

```python
# now if we want to get all the unique dates, we will use a set
sales_set = set()
for sale in sales[1:]:
    sales_set.add(sale.split(',')[1])
```

```python
# we can also perform the one liner 
sales_set = {sale.split(',')[1] for sale in sales[1:]}
```

```python
import datetime as dt
def is_weekend(sale_date):
    return dt.datetime.strptime(sale_date, '%Y-%m-%d').weekday() in (5, 6)
```
