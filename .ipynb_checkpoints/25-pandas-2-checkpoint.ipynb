```python
## series is just like list, but it can also be formed from a dictionary
```

```python
d = {"JAN": 10, "FEB": 15, "MAR": 12, "APR": 16}
```

```python
d['JAN']
```

```python
import pandas as pd
s = pd.Series(d)
```

```python
s['JAN']
```

```python
s[0]
```

```python
s
```

```python
import pandas as pd
s = pd.Series(d, name='val')
```

```python
s
```

```python
s[1:3]
```

```python
s.sum()
```
