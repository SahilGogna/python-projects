<h2>Task 1</h2>
<ul>
    <li>Use orders and get total number of records for a given month (201401).</li>
    <li>Develop a function which take orders collection and month as arguments.</li>
    <li>Month will be passed as integer in the form of yyyyMM (example 201401).</li>
    <li>Return the order count</li>
</ul>

```python
def getMonthlyOrders(orders,month):
    count = 0
    for order in orders:
        date = int(order.split(',')[1][:7].replace('-',''))
        if date == month:
            count += 1
    return count
```

```python
getMonthlyOrders(orders, 201401)
```

<h2>Task 3</h2>
<ul>
    <li>
Use order items data set and get total number of items sold as well as total revenue generated for a given product_id.</li>
<li>Define a function which takes order_items and a product_id as arguments.</li>
<li>product_id will be passed as integer</li>
<li>Get number of items sold for a given product id using quantity (4th field)</li>
<li>Compute revenue generated for a given product id using subtotal (5th field)</li>
<li>Return the number of items sold as well as revenue generated. When we return more than one value, then the type of the returned results will be tuple.</li>
</ul>

```python
def sale_rev(order_items, productId):
    retTup = (0,0.0)
    for item in order_items:
        if productId == int(item.split(',')[2]):
            sale = int(item.split(',')[3])
            rev = float(item.split(',')[4])
            retTup = (retTup[0]+sale,retTup[1]+rev)
    return retTup
```

```python
sale_rev(order_items, 502)
```
